```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
dataset = np.random.randint(0, 100, [50, 5])

dataset
```

```python
_, subs = plt.subplots(1, dataset.shape[1])

for p, i in zip(subs, range(dataset.shape[1])):
    p.set_title(f"Col {i+1}")
    p.plot(np.arange(0, dataset.shape[0]), dataset[:, i])
```

```python
dataset = dataset.astype(np.float16)

dataset = (dataset - dataset.mean(axis=0)) / dataset.std(axis=0)

dataset
```

```python
_, subs = plt.subplots(1, dataset.shape[1])

for p, i in zip(subs, range(dataset.shape[1])):
    p.set_title(f"Col {i+1}")
    p.plot(np.arange(0, dataset.shape[0]), dataset[:, i])
```
