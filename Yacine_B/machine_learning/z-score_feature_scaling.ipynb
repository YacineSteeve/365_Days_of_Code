```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
dataset = np.random.randint(0, 100, [50, 5])

dataset
```

```python
def see_data(d):
    plt.figure(figsize=(15, 5))

    for i in range(d.shape[1]):
        plt.subplot(1, d.shape[1], i+1)
        plt.plot(np.arange(0, d.shape[0]), d[:, i])
        plt.title(f"Col {i+1}")

see_data(dataset)
```

```python
dataset = dataset.astype(np.float16)

dataset = (dataset - dataset.mean(axis=0)) / dataset.std(axis=0)

dataset
```

```python
see_data(dataset)
```
