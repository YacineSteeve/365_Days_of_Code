```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
```

```python
image = plt.imread('bacterias.jpeg')
```

```python
image.shape
```

```python
image = image[:, :, 0]
image.reshape(image.shape[0], image.shape[1], 1)
image.shape
```

```python
plt.figure(figsize=(12, 6))
plt.imshow(image, cmap='gray')
plt.title("Original image")
plt.show()
```

```python
img = np.copy(image)

plt.figure(figsize=(12, 6))
plt.hist(img.ravel(), bins=255)
plt.xlabel("Pixel color")
plt.ylabel("Number")
plt.title("Color repartition")
plt.show()
```

```python
image = image >=115

plt.figure(figsize=(12, 6))
plt.imshow(image)
plt.title("No-background image")
plt.show()
```

```python
cleaned_image = ndimage.binary_opening(image)

plt.figure(figsize=(12, 6))
plt.imshow(image)
plt.title("Opened image")
plt.show()
```

```python
label_image, n_labels = ndimage.label(image)
n_labels
```

```python
plt.figure(figsize=(12, 6))
plt.imshow(label_image)
plt.title("Labeled image")
plt.show()
```

```python
sizes = ndimage.sum(image, label_image, range(n_labels))

plt.figure(figsize=(12, 6))
plt.scatter(range(1, n_labels+1), sizes)
plt.xlabel("Label n°")
plt.ylabel("Size")
plt.title("Labels sizes")
plt.show()
```

```python
sizes = sizes[sizes >= 50]

plt.figure(figsize=(12, 6))
plt.scatter(range(1, len(sizes)+1), sizes)
plt.xlabel("Label n°")
plt.ylabel("Size")
plt.title("Labels sizes")
plt.show()
```
